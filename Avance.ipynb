```python
import pandas as pd

# Intentar cargar el dataset ignorando las l√≠neas con errores
df = pd.read_csv('amazon1.csv', on_bad_lines='skip')

print("Datos cargados exitosamente:")
print(df.head())
```

```python
resumen = df.describe()
print("Resumen de los datos:\n", resumen)
```

```python
tipos_datos = df.dtypes
print("\nTipos de datos:\n", tipos_datos)
```

```python
dimensiones = df.shape
print("\nDimensiones del dataset:", dimensiones)
```

```python
primeras_filas = df.head()
print("\nPrimeras filas del dataset:\n", primeras_filas)
```

```python
cantidad_por_clase = df['rating'].value_counts()
print("\nCantidad de datos por rating:\n", cantidad_por_clase)
```

```python
distribucion_clases = df['actual_price'].value_counts(normalize=True)
print("\nDistribuci√≥n por precio actual:\n", distribucion_clases)
```

```python
def correlation():
    # Seleccionar solo las columnas de tipo float
    float_df = df.select_dtypes(include=[float])
    # Calcular la correlaci√≥n usando el m√©todo de Pearson
    correlations = float_df.corr(method='pearson')
    print("Correlaciones entre columnas de tipo float:\n", correlations)

# Llamar a la funci√≥n
correlation()
```
